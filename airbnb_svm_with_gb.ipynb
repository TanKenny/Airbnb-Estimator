```python
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import codecs, json 

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./dataset/airbnb_seasons_20172018/airbnb_cleaned_fall.csv', encoding='Latin-1')
```

```python
df['neighbourhood_cleansed'] = df['neighbourhood_cleansed'].apply(lambda x: x.lower())
```

```python
df['neighbourhood_cleansed'] = df['neighbourhood_cleansed'].replace(' ', '')
df['neighbourhood_cleansed'] = df['neighbourhood_cleansed'].str.replace(r"[\"\',]", '')
df['neighbourhood_cleansed'] = df['neighbourhood_cleansed'].str.replace("-", "")

df['neighbourhood_cleansed'].value_counts()
```

```python
list = np.unique(df['neighbourhood_cleansed'])
list = list.tolist()
```
